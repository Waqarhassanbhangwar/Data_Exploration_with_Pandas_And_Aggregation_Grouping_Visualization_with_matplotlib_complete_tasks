```python
#Group data by a categorical variable and calculate the mean, median, or other statistics for each group.
import pandas as pd

# Read data from CSV file
df = pd.read_csv('Details.csv')  
df
```

```python
# Grouping data by 'Category' and calculating mean and median
grouped_data = df.groupby('Category')['Quantity'].agg(['mean', 'median'])
print(grouped_data)
```

```python
#Aggregate data on a time series dataset by daily, weekly, or monthly averages.

import pandas as pd

# Create a sample DataFrame with a datetime index
date_rng = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = range(1, len(df) + 1)

# Set 'date' column as the index
df.set_index('date', inplace=True)

# Aggregate data by daily averages
daily_avg = df.resample('D').mean()

# Aggregate data by weekly averages
weekly_avg = df.resample('W').mean()

# Aggregate data by monthly averages
monthly_avg = df.resample('M').mean()

print("Daily Averages:")
print(daily_avg.head())
print("\nWeekly Averages:")
print(weekly_avg.head())
print("\nMonthly Averages:")
print(monthly_avg.head())
```
